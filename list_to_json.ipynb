```python
import pickle
import json
from data_utils.utils import Config
from data_utils.vocab_tokenizer import *
from data_utils.pad_sequence import keras_pad_fn
```

```python
with open("./bert1_aihub/vocab_aihub1.pkl", 'rb') as f:
    vocab = pickle.load(f)

token_list=[]
for i in range (0,23945):
    token_list.append(vocab.idx_to_token[i])

dictionary=dict()
dictionary["idx_to_token"]=token_list
dictionary["token_to_idx"]=vocab.token_to_idx

dictionary["reserved_tokens"]=["[MASK]", "[SEP]", "[CLS]"]
dictionary["unknown_token"]="[UNK]"
dictionary["padding_token"]="[PAD]"
dictionary["bos_token"]=None
dictionary["eos_token"]=None
dictionary["mask_token"]="[MASK]"
dictionary["sep_token"]="[SEP]"
dictionary["cls_token"]="[CLS]"

f.close()

with open("./bert1_aihub/jvocab_aihub1.json", 'w', encoding='utf-8') as j:
    json.dump(dictionary,j, ensure_ascii=False)
j.close()
```
